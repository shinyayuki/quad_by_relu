```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot

def relu(x):
    return np.maximum(0, x)

x = np.linspace(-3., 3., 300)
y = relu(-x) + relu(x)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(-x-1)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(-x-1) + relu(-x) + relu(x)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(x-1)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(-x-1) + relu(-x) + relu(x) + 2*relu(x-1)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(-x-2)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(-x-2) + 2*relu(-x-1) + relu(-x) + relu(x) + 2*relu(x-1)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(x-2)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```

```python
y = 2*relu(-x-2) + 2*relu(-x-1) + relu(-x) + relu(x) + 2*relu(x-1) + 2*relu(x-2)

pyplot.xlim(-3.5, 3.5)
pyplot.ylim(-1, 6)
pyplot.plot(x, y)
```
